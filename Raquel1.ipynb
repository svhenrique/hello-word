```
# Bibliotécas Q12.2 - Q12.3
from matplotlib import pyplot as plt
import math
```

## Gráfico PMF 

```
p = 0.2 # sucesso
def pmf(*valores):
  resultados = []
  for x in valores[0]:
    resultados.append(((1-p)**(x-1))*p)
  return resultados

# Gráfico PMF dos 20 primeiros números do conjunto dos naturais (sem o 0)
# Como na questão X -> infinito, foram selecionados paenas 20 números para o gráfico 
# porém, existe claramente um limite que tende a 0, pois, quanto maior o x mais perto de 0 
# se torna o valor de y.

lista = list(range(1,21))

plt.figure(figsize=(12.5,9)) 
plt.ticklabel_format(useOffset=False, style='plain')
plt.xticks(range(1,21))
plt.title('PMF')

plt.plot(lista, pmf(lista))
plt.show()
```

## Gráfico CDF

```
def cdf(valor):
  probabilidade = pmf([valor]) #lembre-se da função pmf
  for valor_anterior in range(valor-1, 0, -1):
    probabilidade[0] += pmf([valor_anterior])[0]
  return probabilidade[0]

# Gráfico CDF dos 20 primeiros números dos conjuntos dos naturais (sem o 0)
# Observe que o gráfico tende a 1 quando x -> infinito. Isso se dá ao fato de que,
# no CDF, o ultimo número aplicado na função será a soma de sua probabilidade e de 
# todas as probabilidades anteriores, que totaliza 1.

lista = list(range(1,21))
valores_pmf = pmf(lista)
valores_cdf = [cdf(valor) for valor in lista]

plt.figure(figsize=(12.5,9)) 
plt.ticklabel_format(useOffset=False, style='plain')
plt.xticks(range(1,21))
plt.title('CDF')

plt.plot(lista, valores_cdf)
plt.show()

```

## Gráfico PMF 

```
def mf_poisson(*valores):
  
  resultados = []

  media = 8

  for x in valores[0]:
    numerador = (media**x)*math.e**(-1*media)
    denominador = 1
    for valor in range(x, 0, -1): # fatorial
      denominador *= valor
    resultados.append(numerador/denominador)
  return resultados

# Gráfico PMF 20 primeiros números do conjunto dos naturais
# Esse gráfico segue uma função de poisson só que com algumas alterações
# feitas pela questão 12.3

lista = list(range(0,21))

plt.figure(figsize=(12.5,9)) 
plt.ticklabel_format(useOffset=False, style='plain')
plt.xticks(range(0,21))
plt.title('PMF')

plt.plot(lista, mf_poisson(lista))
plt.show()
```

```
def cdf(valor):
  probabilidade = mf_poisson([valor]) #lembre-se da função pmf
  for valor_anterior in range(valor-1, -1, -1):
    probabilidade[0] += mf_poisson([valor_anterior])[0]
  return probabilidade[0]

# Gráfico CDF 20 primeiros números do conjunto dos naturais

lista = list(range(0,21))
valores_cdf = [cdf(valor) for valor in lista]

plt.figure(figsize=(12.5,9)) 
plt.ticklabel_format(useOffset=False, style='plain')
plt.xticks(range(0,21))
plt.title('CDF')

plt.plot(lista, valores_cdf)
plt.show()
```
